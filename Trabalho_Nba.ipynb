# **NBA Project for Classroom  of Introduction a Data Science**

NBA nada mais é que a sigla para National Basketball Association, ou, em tradução livre, Associação Nacional de Basquete. Trata-se da principal organização do esporte nos Estados Unidos, responsável pela maior liga de basquete do mundo, que também leva o nome da associação.

A liga reúne os maiores times de basquete, também conhecidos como “franquias”, dos Estados Unidos. Mas também aceita algumas equipes do Canadá. As equipes se dividem em regionais e se enfrentam em uma série de disputas divididas em fases como a pré-temporada, a temporada regular e os playoffs. Ao fim, um time se consagra campeão.

Este projeto tem como objetivo analisar o desempenho dos jogadores da NBA na temporada 2023-2024, utilizando dados estatísticos extraídos da API disponível no Kaggle (https://www.kaggle.com/datasets/vivovinco/2023-2024-nba-player-stats). A NBA é uma das ligas esportivas mais competitivas do mundo, e a análise de métricas individuais permite identificar quais atletas tiveram maior impacto durante a temporada regular.

A partir dos dados coletados, serão avaliados diversos aspectos do desempenho dos jogadores, como pontuação em arremessos de 3 pontos e 2 pontos, aproveitamento de arremessos de quadra (Field Goal Percentage), lances livres (Free Throw Percentage), rebotes e assistências. Com essas informações, buscamos determinar quais atletas foram mais eficientes ao longo da temporada regular, considerando diferentes métricas de desempenho.

# **Desenvolvimento**

**Arremessos de 3 pontos (3P%)** - 3P,3PA,3P%,Player,Pos - Top 15 Player 3 pontos

##3 pontos##

```python
# Tabela 3 pontos
tres_pontos = tabela[['Player','Pos','3P','3PA','3P%']]
```

```python
# top 15 arremessos convertidos
tres_pontos.sort_values(['3P'],ascending=False).head(15).reset_index()
```

```python
# Top 15 mais possuem tentativas de arremos de 3 pontos
tres_pontos.sort_values(['3PA'],ascending=False).head(15).reset_index()
```

```python
from pickle import TRUE
# Aqui eu ordeno dos top 15 principais, que mais possuem porcentagem de arremos convertidos, baseando na % e tentativas
principais = tres_pontos.sort_values(['3P','3PA'],ascending=False).head(15).reset_index()
principais
```

Isso indica quantos dos arremessos foram tentados no perímetro e mostra quão focada uma equipe está em arremessar de longe ou de perto. A média da NBA nos arremessos de longe fica próxima a 35%

```python
media_arremessos_3pontos = round(tres_pontos['3P%'].mean()*100)
print(f'Media de arremessos total convertidos em porcentagem: {media_arremessos_3pontos}%')
```

```python
top_5_grafico_3pontos =  tres_pontos.sort_values(['3P','3PA'],ascending=False).head(5).reset_index()
top_5_grafico_3pontos
```

```python
plt.figure(figsize=(14, 7))
# Barras para pontuações médias
bars1 = plt.bar(top_5_grafico_3pontos['Player'], top_5_grafico_3pontos['3P'], width=0.5, label='Média por jogo', color='#920f09', align='center')
# Barras para cestas de 3 pontos
bars2 = plt.bar(top_5_grafico_3pontos['Player'], top_5_grafico_3pontos['3PA'], width=0.3, label='Tentativas de 3 pontos', color='#ce2e1a', align='edge')

# Adicionando valores no topo de cada barra
for bar in bars1:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/4, height, f'{height:.1f}', ha='center', va='bottom', fontsize=14, color='#920f09')

for bar in bars2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{height:.1f}', ha='center', va='bottom', fontsize=14, color='#ce2e1a')

for i, player in enumerate(top_5_grafico_3pontos['Player']):
    plt.text(i, -0.7, str(i+1)+"º", ha='center', va='top', fontsize=14, color='black')

plt.title('Top 5 (Jogadores Cestas Tentadas e Convertidas - 3P)')
plt.ylabel('Cestas e 3P%  Convertidas (média por jogo)')
plt.xticks(rotation=0, ha="center",fontweight='bold')
plt.legend()
plt.ylim(0, 13)
plt.grid(True, axis='y',alpha=0.5)
plt.ylim(0, 14)
plt.yticks(range(0, 14, 1))
plt.tick_params(axis='x', labelsize=12)
plt.tight_layout()
plt.show()
```

É a porcentagem de cestas feitas com assistências de um jogador enquanto ele está em quadra. É o passe que se torna uma cesta em seguida. Times em que apenas um jogador possui assistências mostra um esquema tático dependente desse jogador para criar espaços e acionar as jogadas. Times com vários jogadores com assistências mostra um jogo mais coletivo, em que a bola gira pela quadra passando pela mão de vários jogadores diferentes em busca de espaço para um arremesso. Individualmente, armadores principais têm em média 4 assistências por jogo, com todas as outras posições ficando próximas a 1.5 assistências.
